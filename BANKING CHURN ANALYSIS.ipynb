# DATA CLEANING

#THE MISSING VALUES IN THE'AGE' COLUMN CAN BE FILLED BY FINDING THE MEAN FOR THE 'AGE' COLUMN:

```python
#To find the mean for the 'Age' column:

df_mean=df['Age'].mean()

#'Age'column is continuous ,so should convert float value to integer value:

df_mean=int(df_mean)
df_mean
```

```python
#Filling the null values in'Age' column by 'MEAN':

df['Age_new']=df['Age'].fillna(df_mean)
```

#THE MISSING VALUES IN THE'GENDER' COLUMN CAN BE FILLED BY FINDING THE MODE FOR THE 'GENDER' COLUMN:

```python
#To find the mode for the 'Gender' column:

df['Gender'].value_counts().index[0]
```

```python
#Filling the null values in 'Gender'  column by MODE:

df['Gender_new']=df['Gender'].fillna(df['Gender'].value_counts().index[0])
df['Gender_new']
```

```python
df.head()
```

```python
#Dropping "Age","GENDER" columns :

df=df.drop(['Age','Gender'],axis=1)
df.head()
```

```python
# GENERAL FOR LOOP FOR FINDING CATEGORICAL AND NUMERICAL COLUMNS:

cat=[]
num=[]

for column in df.columns:
    if df[column].nunique()> 10:
        num.append(column)
    else:
        cat.append(column)
```

```python
# Adding labels to age bins for analysis:

labels=['0-20','21-40','41-60','61-80','Above 80']
bins=[0,20,40,60,80,100]
df['Age_bins']=pd.cut(df.Age_new,bins,labels=labels,include_lowest=True)
df.head()
```

```python
#Bar chart
plt.bar(labels,df["Age_bins"].value_counts(),label="Age bins")
plt.xlabel("Age")
plt.ylabel("Num of Customers")
plt.title("Churn rate of Customers",fontsize=15)
plt.legend(loc="best")
plt.show()
```

Numeric Analysis:

1.The bank has a high rate of customers between the age 0-20.
2.The bank has a lowest rate of customers which are above the age 80.
